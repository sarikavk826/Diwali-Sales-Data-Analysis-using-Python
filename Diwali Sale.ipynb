# Checking the Outliers

```python
column_with_outliers = ['Age', 'Orders','Amount']

for col in column_with_outliers:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3-Q1
    lb = Q1-1.5*IQR
    ub = Q3+1.5*IQR

    median = df[col].median()
    outlier = (df[col]<lb) | (df[col]>ub)
    print(outlier.sum())
```

```python
df['Gender'].value_counts() # checking in the gender who made the most purchased by using count
```

```python
ax = sns.countplot(data = df, x = 'Gender', palette = 'Set2')

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
#  which gender spends the more money (using reset index craeting dataframe
gen_sales = df.groupby('Gender')['Amount'].sum().reset_index()
gen_sales
```

```python
t = df.groupby('Gender')['Amount'].sum().plot(kind = 'bar', color=['pink', 'lightblue']) # here is showing female spend the more money on shopping
for bar in t.containers:
    t.bar_label(bar)
plt.title("Total Shopping Amount by Gender")
plt.xlabel("Gender")
plt.ylabel("Total Amount Spent")
plt.show()

```

```python
ad = sns.countplot(data = df, x = 'Age Group', palette = 'Set2') # which age group is purchasing more
ad
# 26-35 age group made the most purchase
```

```python
ad = sns.countplot(data = df, x = 'Age Group', hue = 'Gender' ,palette = 'Set2') # which age group is purchasing more dividing by gender

for bar in ad.containers:
    ad.bar_label(bar)

# 26-35 age group made the most purchase specially female
```

```python
# total amount vs age group
colors = sns.color_palette("Set2")

df.groupby('Age Group')['Amount'].sum().sort_values(ascending = False).plot(kind = 'bar', color = colors)
plt.show()

# 26-35 age group spend more money on shoppings
```

```python
colors = sns.color_palette("Set2")
df.groupby('State')['Orders'].sum().sort_values(ascending = False).plot(kind = 'bar', color = colors)
plt.show
```
